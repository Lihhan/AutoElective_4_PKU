2. 配置个人信息（您输入的信息不会被任何人知道）

```python
import ipywidgets as widgets
from IPython.display import display

def generate_config(student_id, password, dual_degree, identity, course_name, course_class, course_school):
    config_content = f"""
# filename: config.ini
# coding: utf-8

[user]
student_id = {student_id}
password = {password}
dual_degree = {dual_degree}
identity = {identity}

[client]
supply_cancel_page = 1
refresh_interval = 8
random_deviation = 0.2
iaaa_client_timeout = 30
elective_client_timeout = 60
elective_client_pool_size = 2
elective_client_max_life = 600
login_loop_interval = 2
print_mutex_rules = true
debug_print_request = false
debug_dump_request = false

[monitor]
host = 127.0.0.1
port = 7074

[course:1]
name = {course_name}
class = {course_class}
school = {course_school}
"""
    config_file_path = "./config.ini"
    with open(config_file_path, 'w') as f:
        f.write(config_content)
    print(f"配置个人信息已完成！")
def on_submit(b):
    student_id = text_student_id.value
    password = text_password.value
    dual_degree = bool(dropdown_dual_degree.value)
    identity = text_identity.value
    course_name = text_course_name.value
    course_class = int(text_course_class.value)
    course_school = text_course_school.value
    if not student_id or not password or not identity or not course_name or not course_school:
        print("所有字段都不能为空！")
        return
    generate_config(student_id, password, dual_degree, identity, course_name, course_class, course_school)
text_student_id = widgets.Text(description="学号:", layout=widgets.Layout(width='400px', height='30px'),
                                style={'description_width': 'initial'})
text_password = widgets.Password(description="密码:", layout=widgets.Layout(width='400px', height='30px'),
                                style={'description_width': 'initial'})
dropdown_dual_degree = widgets.Dropdown(description="是否为双学位账号:", options=[True, False],
                                        layout=widgets.Layout(width='400px', height='30px'),
                                        style={'description_width': 'initial'})
text_identity = widgets.Text(description="(bzx,bfx) 对应于主修/辅双:", options=['bzx', 'bfx'],
                            layout=widgets.Layout(width='400px', height='30px'),
                            style={'description_width': 'initial'})
text_course_name = widgets.Text(description="课程名:", layout=widgets.Layout(width='400px', height='30px'),
                                style={'description_width': 'initial'})
text_course_class = widgets.IntText(description="班号:", layout=widgets.Layout(width='400px', height='30px'),
                                style={'description_width': 'initial'})
text_course_school = widgets.Text(description="开课单位:", layout=widgets.Layout(width='400px', height='30px'),
                                style={'description_width': 'initial'})
submit_button = widgets.Button(description="确认",button_style='success',
                    layout=widgets.Layout(width='80px', height='30px'),
                    style={'font_weight': 'bold'}
                )
submit_button.on_click(on_submit)
display(text_student_id, text_password, dropdown_dual_degree, text_identity, 
        text_course_name, text_course_class, text_course_school, submit_button)
```

3. 为了避免用户配置pytorch环境的麻烦，需要您在tt识图(http://www.ttshitu.com/)中注册一个账号并将账号密码填入下列框中。

```python
import json
import ipywidgets as widgets
def generate_apikey(username, password):
    api_key_content = {
        "username": username,
        "password": password
    }
    with open('apikey.json', 'w') as f:
        json.dump(api_key_content, f, indent=4)
    print("该步骤已完成！")
def on_generate_apikey(b):
    username = username_input.value
    password = password_input.value
    if not username or not password:
        print("用户名和密码不能为空！")
        return
    generate_apikey(username, password)
username_input = widgets.Text(
    description="用户名:", 
    layout=widgets.Layout(width='400px', height='30px'),
    style={'description_width': 'initial'}
)
password_input = widgets.Password(
    description="密码:", 
    layout=widgets.Layout(width='400px', height='30px'),
    style={'description_width': 'initial'}
)
generate_button = widgets.Button(
    description="生成 apikey.json",
    button_style='success',
    layout=widgets.Layout(width='150px', height='40px'),
    style={'font_weight': 'bold'}
)
generate_button.on_click(on_generate_apikey)
display(username_input, password_input, generate_button)
```
